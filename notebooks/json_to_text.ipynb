### Purpose:
converts JSON format into normalized YOLO TXT format.
(For Every Single Image !)


```python
# For simplicity, I will start with one json file (data/Car damages dataset/File1/ann/Car damages 2.jpg.json)
#then work my way up to all files

import os 
print(f'current directory{os.getcwd()}')

path = "../data/Car damages dataset/File1/ann/Car damages 2.jpg.json"


# Load the file using json module
import json

if os.path.exists(path):
    with open(path , "r") as file: 
        data = json.load(file) 
        
        #the json file has many tags, let's get the size tags since that's what's important
        print(f'Image size is: {data["size"]}')


else:
    print(f"Can't find the file, check the path {os.getcwd()}")
```

### Cool, Now we need to start Normalizing:

Loop through every object in `data['objects']`.

Inside that, loop through every point in `object['points']['exterior']`.

Calculate: `x_norm​=point[0]/width` and `y_norm​=point[1]/height`

```python
# YOLO list to store the final lines for the .txt file
# (This is for all objects !)
yolo_lines = []


#get the height and width seperatly
img_width = data["size"]["width"]
img_height = data["size"]["height"]


#start the loop for each object
for each_object in data['objects']:

    #get the id (for later remapping !!)
    class_id = each_object['classId']
    
    # the part id (hood , bumper ..etc)
    part_id = each_object['id']

    #The class name (maybe useful later)
    class_name = each_object["classTitle"]


    #make a list specifically for this object
    current_object_points = [str(class_id)] #Put the class id in the beggining (as a string !!!) becuase yolo wants it to be string


    #start Normalizing: 
    for coordinates in each_object["points"]['exterior']:

        x_norm = round(coordinates[0] / img_width , 7)
        y_norm = round(coordinates[1] / img_height , 7)

        current_object_points.append(str(x_norm))   #Yolo wants this to be str
        current_object_points.append(str(y_norm))   #also this is str
        
        # Now we need to convert the list from ['1', '0.5', '0.6'] to '1' '0.5' '0.6' 
        single_yolo_lines = " ".join(current_object_points)

        #End of Normalization loop

    yolo_lines.append(single_yolo_lines)
    #end of each_object loop

```
